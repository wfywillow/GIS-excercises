```python
import os

import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.geometry import LineString
```

```python
os.chdir(r'C:\Users\BEIJING-c025\OneDrive\Materials\GIS\practice\practice_1')
```

```python
crs_init='epsg:4326'
crs_out='epsg:26718'
```

```python
df=pd.read_csv(r'202011-citibike-tripdata.csv')
df
```

```python
lines=df.apply(lambda row: LineString([(row['start station longitude'],row['start station latitude']),(row['end station longitude'],row['end station latitude'])]),axis=1,)
```

```python
lines.head()
```

```python
gdf=gpd.GeoDataFrame(df,geometry=gpd.GeoSeries(lines),crs=crs_init)
gdf
```

```python
gdf=gdf.to_crs(crs_out)
gdf.crs
```

```python
gdf['linedistance']=gdf['geometry'].length
gdf
```

```python
gdf.rename(columns={"tripduration":"duration",
                            "start station id":"startid", 
                            "start station name":"startname",
                            "start station latitude":"start_lat",
                            "start station longitude":"start_lon",
                            "end station id":"endid",
                            "end station name":"endname",
                            "end station latitude":"end_lat",
                            "end station longitude":"end_lon",
                            "birth year":"birthyear",
                            "linedistance":"linedist"},inplace=True)
```

```python
assert np.all(gdf.columns.str.len() <= 10), \
"Column names will be truncated when saving as shapefile!"
```

```python
os.makedirs('output')
gdf.to_file(os.path.join('output','biketrip.shp'))
```
